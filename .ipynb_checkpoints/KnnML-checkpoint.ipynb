```python
## KNN Classification Algorithm (K = hyperparameters => set by us humans)
### Predict class (label) based on buying,maint,door,persons,lug_boot,safety features

from sklearn import preprocessing
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
data = pd.read_csv("car.data", sep=",")
data.head()
```

```python
##coverting non numerical to numeric : Getting data to correct form

le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))

```

```python
X = list(zip(buying, maint, door, persons, lug_boot, safety))  # features (datapoints)
Y = list(cls)  # labels (predict)
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size = 0.2)
```

```python
model = KNeighborsClassifier(n_neighbors=9) #hyperparameter (highest acc = 9)
model.fit(x_train, y_train)
```

```python
acc = model.score(x_test, y_test)
print(acc)
```

```python
predicted = model.predict(x_test)
names = ["unacc", "acc", "good", "vgood"]

for x in range(len(predicted)):
    print("Predicted: ", names[predicted[x]], "Data: ", x_test[x], "Actual: ", names[y_test[x]])
```

```python
for x in range(len(predicted)):
    print("Predicted: ", names[predicted[x]], "Data: ", x_test[x], "Actual: ", names[y_test[x]])
    n = model.kneighbors([x_test[x]], 9, True)
    print("N: ", n)
```

```python
##Support Vector Machine
### ADD Z AXIS like => kernel (hyperplane) x1,x2,x3
import sklearn
from sklearn import svm
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer() ##get data from sklearn lib
```

```python
print("Features: ", cancer.feature_names)
print("Labels: ", cancer.target_names)
```

```python
X = cancer.data #features
Y = cancer.target #labels

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X,Y,test_size = 0.2)
```

```python
print("X train val:" , x_train)
print("Y train val:" , y_train)
```

```python
from sklearn import metrics
classes = ["malignant", "benign"]

clf = svm.SVC(kernel = "linear", C=2)
clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)

acc = metrics.accuracy_score(y_test,y_pred)
print(acc)
```

```python
clf2 = KNeighborsClassifier(n_neighbors=9)
clf2.fit(x_train,y_train)

y_pred = clf2.predict(x_test)

acc2 = metrics.accuracy_score(y_test,y_pred)
print(acc2)
```

```python
diff = acc - acc2
diffP = diff * 100
print("svm difference (margin) with KNN: ", diffP , "%")
```
